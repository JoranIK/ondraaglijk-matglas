```python
# Imports
import re
import os
import numpy as np
import pandas as pd
import pickle
```

```python
def ministerie_name_MG(s):
    if s == 'LNV':
        return 'Ministerie Van Landbouw, Natuur En Voedselkwaliteit'
    elif s == 'AZ':
        return 'Ministerie Van Algemene Zaken'
    elif s == 'BuZa':
        return 'Ministerie Van Buitenlandse Zaken'
    elif s == 'BZK':
        return 'Ministerie Van Binnenlandse Zaken En Koninkrijksrelaties'
    elif s == 'EZK':
        return 'Ministerie Van Economische Zaken En Klimaat'
    elif s == 'Financia╠én':
        return 'Ministerie Van Financiën'
    elif s == 'I&W':
        return 'Ministerie Van Infrastructuur En Waterstaat'
    elif s == 'J&V':
        return 'Ministerie Van Justitie En Veiligheid'
    elif s == 'Def':
        return 'Ministerie Van Defensie'
    elif s == 'OCW':
        return 'Ministerie Van Onderwijs, Cultuur En Wetenschap'
    elif s == 'SZW':
        return 'Ministerie Van Sociale Zaken En Werkgelegenheid'
    elif s == 'VWS':
        return 'Ministerie Van Volksgezondheid, Welzijn En Sport'
    
def ministerie_name_OT(s):
    if s == 'LNV':
        return 'Ministerie Van Landbouw, Natuur En Voedselkwaliteit'
    elif s == 'AZ':
        return 'Ministerie Van Algemene Zaken'
    elif s == 'BuZa':
        return 'Ministerie Van Buitenlandse Zaken'
    elif s == 'BZK':
        return 'Ministerie Van Binnenlandse Zaken En Koninkrijksrelaties'
    elif s == 'EZK':
        return 'Ministerie Van Economische Zaken En Klimaat'
    elif s == 'FIN':
        return 'Ministerie Van Financiën'
    elif s == 'IenW':
        return 'Ministerie Van Infrastructuur En Waterstaat'
    elif s == 'JenV':
        return 'Ministerie Van Justitie En Veiligheid'
    elif s == 'Defensie':
        return 'Ministerie Van Defensie'
    elif s == 'OCW':
        return 'Ministerie Van Onderwijs, Cultuur En Wetenschap'
    elif s == 'SZW':
        return 'Ministerie Van Sociale Zaken En Werkgelegenheid'
    elif s == 'VWS':
        return 'Ministerie Van Volksgezondheid, Welzijn En Sport'
```

```python
excels_path_OT = "data/ExcelToDF/excel/Ondraaglijk_traag/"

excel_dict_OT = {}  
for excel_OT in os.listdir(excels_path_OT):
    ministerie_OT = re.search(r"[A-Za-z]+(?=\.)", str(excel_OT))[0]
    df_OT = pd.read_excel(excels_path_OT + excel_OT)
    df_OT['Ministerie'] = ministerie_name_OT(ministerie_OT)
    excel_dict_OT[ministerie_OT] = df_OT

LNV_OT = excel_dict_OT['LNV']
AZ_OT = excel_dict_OT['AZ']

# in de AZ excel waren een aantal lege rijen waar de excel formules wel een 0 neerzetten bij de doorlooptijd
# er waren geen rijen waar de echte doorlooptijd 0 dagen was, dus door te filteren op doorlooptijden hoger dan nul
# worden geen 'echte' doorlooptijden verwijderd, alleen foute
AZ_OT = AZ_OT[AZ_OT['Aantal dagen \nin behandeling'] > 0]

BuZa_OT = excel_dict_OT['BuZa']
BZK_OT = excel_dict_OT['BZK']
EZK_OT = excel_dict_OT['EZK']
FIN_OT = excel_dict_OT['FIN']
IenW_OT = excel_dict_OT['IenW']
JenV_OT = excel_dict_OT['JenV']
OCW_OT = excel_dict_OT['OCW']
SZW_OT = excel_dict_OT['SZW']
VWS_OT = excel_dict_OT['VWS']
DEF_OT = excel_dict_OT['Defensie']

sheetMega_OT = pd.concat([LNV_OT, 
                          AZ_OT, 
                          BuZa_OT, 
                          BZK_OT, 
                          EZK_OT, 
                          FIN_OT, 
                          IenW_OT, 
                          JenV_OT, 
                          OCW_OT, 
                          SZW_OT, 
                          VWS_OT, 
                          DEF_OT], axis=0) 
```

```python
excels_path_MG = "data/ExcelToDF/excel/Matglas/"

excel_dict_MG = {}
for excel_MG in os.listdir(excels_path_MG):
    ministerie_MG = re.search(r"^[^,]*", str(excel_MG))[0]
    df_MG = pd.read_excel(excels_path_MG + excel_MG)
    df_MG['Ministerie'] = ministerie_name_MG(ministerie_MG)
    excel_dict_MG[ministerie_MG] = df_MG
    
AZ_MG = excel_dict_MG['AZ']
BuZa_MG = excel_dict_MG['BuZa']
BZK_MG = excel_dict_MG['BZK']
Corona_MG = excel_dict_MG['Corona']
Def_MG = excel_dict_MG['Def']
EZK_MG = excel_dict_MG['EZK']
Fin_MG = excel_dict_MG['Financia╠én']
IW_MG = excel_dict_MG['I&W']
JV_MG = excel_dict_MG['J&V']
LNV_MG = excel_dict_MG['LNV']
OCW_MG = excel_dict_MG['OCW']
SZW_MG = excel_dict_MG['SZW']
VWS_MG = excel_dict_MG['VWS']

BuZa_MG.rename(columns={"Datum van ontvangst": "Datum van binnenkomst", "Datum van besluit": "Datum van antwoord"}, inplace=True)
OCW_MG.rename(columns={"Datum van ontvangst": "Datum van binnenkomst", "Datum van besluit": "Datum van antwoord"}, inplace=True)
SZW_MG.rename(columns={"Datum van ontvangst": "Datum van binnenkomst", "Datum van besluit": "Datum van antwoord"}, inplace=True)
VWS_MG.rename(columns={"Datum van ontvangst": "Datum van binnenkomst", "Datum van besluit": "Datum van antwoord"}, inplace=True)

sheetMega_MG = pd.concat([AZ_MG, 
                          BuZa_MG, 
                          BZK_MG, 
                          Def_MG, 
                          EZK_MG, 
                          Fin_MG, 
                          IW_MG, 
                          JV_MG, 
                          LNV_MG, 
                          OCW_MG, 
                          SZW_MG, 
                          VWS_MG], axis=0).reset_index(drop=True)
```

```python
excels_path_MG2 = "data/ExcelToDF/excel/Matglas/"

excel_dict_MG2 = {}
for excel_MG2 in os.listdir(excels_path_MG2):
    ministerie_MG2 = re.search(r"^[^,]*", str(excel_MG2))[0]
    df_MG2 = pd.read_excel(excels_path_MG2 + excel_MG2, sheet_name=1)
    df_MG2['Ministerie'] = ministerie_name_MG(ministerie_MG2)
    excel_dict_MG2[ministerie_MG2] = df_MG2
    
AZ_MG2 = excel_dict_MG2['AZ']
BuZa_MG2 = excel_dict_MG2['BuZa']
BZK_MG2 = excel_dict_MG2['BZK']
Corona_MG2 = excel_dict_MG2['Corona']
Def_MG2 = excel_dict_MG2['Def']
EZK_MG2 = excel_dict_MG2['EZK']
Fin_MG2 = excel_dict_MG2['Financia╠én']
IW_MG2 = excel_dict_MG2['I&W']
JV_MG2 = excel_dict_MG2['J&V']
LNV_MG2 = excel_dict_MG2['LNV']
OCW_MG2 = excel_dict_MG2['OCW']
SZW_MG2 = excel_dict_MG2['SZW']
VWS_MG2 = excel_dict_MG2['VWS']

BuZa_MG2.rename(columns={"Datum van ontvangst": "Datum van binnenkomst", "Datum van besluit": "Datum van antwoord"}, inplace=True)
OCW_MG2.rename(columns={"Datum van ontvangst": "Datum van binnenkomst", "Datum van besluit": "Datum van antwoord"}, inplace=True)
SZW_MG2.rename(columns={"Datum van ontvangst": "Datum van binnenkomst", "Datum van besluit": "Datum van antwoord"}, inplace=True)
VWS_MG2.rename(columns={"Datum van ontvangst": "Datum van binnenkomst", "Datum van besluit": "Datum van antwoord"}, inplace=True)

sheetMega_MG2 = pd.concat([AZ_MG2, 
                           BuZa_MG2, 
                           BZK_MG2, 
                           Def_MG2, 
                           EZK_MG2, 
                           Fin_MG2, 
                           IW_MG2, 
                           JV_MG2, 
                           LNV_MG2, 
                           OCW_MG2, 
                           SZW_MG2, 
                           VWS_MG2], axis=0).reset_index(drop=True)
```

```python
sheetMega_MG2['Aantal dagen \nin behandeling'] = sheetMega_MG2.apply(lambda x: pd.to_numeric(x['Aantal dagen \nin behandeling'], errors='coerce'), axis=1)
sheetMega_MG2 = sheetMega_MG2[sheetMega_MG2['Aantal dagen \nin behandeling'].notna()]
```
