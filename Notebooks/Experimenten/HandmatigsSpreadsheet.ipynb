```python
import pandas as pd
import dateparser
```

```python
handmatig = pd.read_csv('data/Handmatig.csv')
handmatig_minder = pd.read_csv('data/HandmatigMinder.csv')

handmatig.set_index('Document ID', inplace=True)
handmatig_minder.set_index('Document ID', inplace=True)

handmatig_minder['Verschil'] = handmatig_minder['Dagtekening'].apply(lambda x: dateparser.parse(x)) - handmatig_minder['Verzoeksdatum'].apply(lambda x: dateparser.parse(x))
handmatig['Verschil'] = handmatig['Dagtekening'].apply(lambda x: dateparser.parse(x)) - handmatig['Verzoeksdatum'].apply(lambda x: dateparser.parse(x))
```
