# Handmatig vs Spacy

### Problemen in het kort:
- Grote getallen als 10053 of 20701: Geen groot probleem, kan niet geparsed worden dus loopt op niks uit. Komt voor in bijna alle documenten. 
- Datums als stempels: Wel een probleem, dit werd niet herkend door de OCR dus staat de datum van het besluit/de dagtekening er niet in. Voorbeeld in Document 1. 
- Getallen met 4 cijfers: Worden herkend als een jaargetal, dit is pas een probleem als ze in de buurt van 2023 zitten. Komt voor in meerdere documenten, voorbeeld in Document 2. 
- Jaargetallen in de kantlijn: Kunnen als ‘eerste’ datum worden geselecteerd. Voorbeeld in Document 3. 
- Het splitsen van datums in datum en jaargetal: Voorbeeld 15 januari 2022 naar 15 januari (2023 wordt dan automatisch toegevoegd) en 1 januari 2022. Voorbeeld in Document 4. 
- OCR fouten: Kunnen tot fouten leiden. Voorbeeld in Document 6. 
- Datums zonder jaargetal: Uit de context is het jaargetal wel duidelijk maar Spacy ziet dit niet. Voorbeeld in Document 7.

```python
# Imports
import re
import spacy
import numpy as np
import pandas as pd
from spacy import displacy
import dateparser
```

```python
woobesluiten = pd.read_csv('data/handmatig/woobesluiten.csv.gz')
```

```python
# multi index
woobesluiten.index
woobesluiten.set_index(['foi_documentId', 'foi_pageNumber'], inplace=True)
```

```python
# alleen ministeries overhouden
wooMinisterie = woobesluiten[woobesluiten['dc_publisher_name'].str.contains('Ministerie')]
wooMinisterie.shape
```

```python
# ministeries van 2023
wooMinisterie['dc_date_year'] = pd.to_datetime(wooMinisterie['dc_date_year'], format='%Y-%m-%d')
ministeries2023 = wooMinisterie[wooMinisterie['dc_date_year'] >= '2023']
```

```python
# 10 (willekeurige) paginas

paginas_10 = ministeries2023[['dc_source', 'dc_source_dossier']].sample(n=10, random_state=1).index.get_level_values(0)
```

```python
# volledige documenten opzoeken

documenten_10 = ministeries2023.loc[(paginas_10, [1, 2, 3]), ]
```

```python
nlp = spacy.load("nl_core_news_lg")
```

```python
# extract dates functie uit notebook van Maarten Marx 'DutchDatumParser'

date_label = ['DATE']

def extract_dates(text):
    doc = nlp(text)
    results = [ent.text for ent in doc.ents if ent.label_ in date_label]
    results= {s:dateparser.parse(s) for s in results}
    return results

def sort_dates(dates):
    if dates != None:
        dates = sorted([x for x in dates.values() if not x is None])
    else:
        dates = dates
    return dates
```

```python
# extract dates

documenten_10['dates'] = documenten_10['foi_bodyTextOCR'].apply(lambda x: extract_dates(x) if (type(x) == str) else None)
```

```python
# datums sorteren

documenten_10['dates_sorted'] = documenten_10['dates'].apply(lambda x: sort_dates(x))
```
